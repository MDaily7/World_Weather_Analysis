```python
# import dependencies
import random
import numpy as np
import timeit 
```

```python
# randint()
random.randint(-90,90)
```

```python
# random.random()
random.random()
```

```python
# combining random.randint() and random.random()
random_number = random.randint(-90,89) + random.random()
random_number
```

```python
# generating 10 random latitudes with the above method
x = 1
latitudes = []
while x < 11:
    random_latitude = random.randint(-90,89) + random.random()
    latitudes.append(random_latitude)
    x+=1
latitudes   
```

```python
# Create latitudes function
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90,90) + random.random()
        latitudes.append(random_lat)
        x+=1
    return latitudes
```

```python
# time latitudes function 
%timeit latitudes(1500)
```
